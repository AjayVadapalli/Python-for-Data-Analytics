```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig, ax = plt.subplots()

# Plot data on the axes
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Show plot
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
plt.plot(x,y)
plt.title("Simple Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y=np.sin(x)
plt.plot(x,y)
plt.title("Line Plot - Sine Wave")  # Adds a title
plt.xlabel("X Axis")         # X-axis label
plt.ylabel("Y Axis")         # Y-axis label
plt.show()                   # Displays the plot
```

```python
categories = ['Apple', 'Banana', 'Cherry', 'Date']
values = [25, 40, 30, 10]
```

```python
plt.bar(categories,values, color=['blue', 'green', 'red', 'purple'])
plt.title("Bar Chart")
plt.ylabel("Quantity")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 30, 25]
```

```python
plt.scatter(x, y, color='r', marker='x', s=100)  # 's' controls size of markers
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Simple Scatter Plot")

plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data,bins=30,color='c',edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Simple Histogram")

plt.show()
```

```python
labels = ["Apples", "Bananas", "Cherries", "Dates"]
sizes = [30, 40, 20, 10]
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['red', 'yellow', 'pink', 'brown'])
plt.title("Fruit Distribution")

plt.show()
```
