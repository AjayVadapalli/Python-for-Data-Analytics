```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("My First Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
plt.grid(True)
```

```python
plt.grid(color='gray', linestyle='--', linewidth=0.5)
```

```python
plt.plot(x, y, label="Product A")
plt.plot(x, y, label="Product B")
plt.legend()
```

```python
plt.plot(x, y, color='red', linestyle='--', marker='o')
```

```python
plt.xlim(1, 4)   # Zoom in on x values from 1 to 4
plt.ylim(10, 30) # Zoom in on y values from 10 to 30
```
